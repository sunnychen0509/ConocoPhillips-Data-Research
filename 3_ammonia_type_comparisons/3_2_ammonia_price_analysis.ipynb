```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("3_1_Ammonia_Price_Chart.csv")
df
```

```python
df_color = df.drop(columns=['region', 'shipping_method'])
grouped_mean = df_color.groupby('color').mean()

grouped_mean
```

```python
# Transpose the DataFrame for easier plotting
df_T = grouped_mean.T

# Create separate line charts for each region
for color in df_T.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(df_T.index, df_T[color], marker='o', color=color, label=color)
    plt.title(f'Monthly Prices for {color}')
    plt.xlabel('Month')
    plt.ylabel('Price')
    plt.legend(title='Color of Hydrogen')
    plt.grid(True)
    if (color == 'blue'):
        plt.yticks(range(440, 460, 2))
    elif (color == 'gray'):
        plt.yticks(range(395, 415, 2))
    else:
        plt.yticks(range(980, 1000, 2))
    plt.show()
```
